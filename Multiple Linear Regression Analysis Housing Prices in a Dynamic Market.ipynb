```python
# Converting the categorical variable into numerical
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
housing[varlist] = housing[varlist].apply(binary_map)

# Check the housing dataframe now
housing
```

```python
# Creating dummy variable
status = pd.get_dummies(housing['furnishingstatus'])

# Check what the dataset 'status' looks like
status
```

```python
# Dropping the first column from status dataset
status = pd.get_dummies(housing['furnishingstatus'], drop_first = True)

# Adding the status to the original housing dataframe
housing = pd.concat([housing, status], axis = 1)

# Dropping 'furnishingstatus' as we have created the dummies for it
housing.drop(['furnishingstatus'], axis = 1, inplace = True)

housing
```

```python
from sklearn.model_selection import train_test_split

# We specify random seed so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(housing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Applying scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train
```

```python
# Dividing the training data set into X and Y
y_train = df_train.pop('price')
X_train = df_train
```

```python
#Build a linear model

import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.summary()
```

```python
# Checking for the VIF values of the variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Creating a dataframe that will contain the names of all the feature variables and their VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python

# Dropping highly correlated variables and insignificant variables
X = X_train.drop('semi-furnished', 1,)

# Build a fitted model after dropping the variable
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()

# Printing the summary of the model
print(lr_2.summary())
```

```python
# Calculating the VIFs again for the new model after dropping semi-furnished

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python

X = X.drop('bedrooms', 1)
# Build a second fitted model
X_train_lm = sm.add_constant(X)
lr_3 = sm.OLS(y_train, X_train_lm).fit()

# Printing the summary of the model
print(lr_3.summary())
```

```python
# Calculating the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X = X.drop('basement', 1)
X_train_lm = sm.add_constant(X)

lr_4 = sm.OLS(y_train, X_train_lm).fit()
print(lr_4.summary())
```

```python
import statsmodels.api as sm
import pandas as pd
import scipy.stats as stats

# Fit the OLS model
lr_4 = sm.OLS(y_train, X_train_lm).fit()

# Calculate the total sum of squares
total_ss = ((y_train - y_train.mean()) ** 2).sum()

# Calculate the residual sum of squares
residual_ss = (lr_4.resid ** 2).sum()

# Calculate the explained sum of squares
explained_ss = total_ss - residual_ss

# Calculate the degrees of freedom
df_total = len(y_train) - 1
df_residual = len(y_train) - len(lr_4.params)

# Calculate the mean squares
ms_explained = explained_ss / (len(lr_4.params) - 1)
ms_residual = residual_ss / df_residual

# Calculate the F-statistic
f_statistic = ms_explained / ms_residual

# Calculate the p-value
p_value = 1 - stats.f.cdf(f_statistic, len(lr_4.params) - 1, df_residual)

# Create the ANOVA table as a DataFrame
anova_data = {'Source of Variation': ['Regression', 'Residual', 'Total'],
              'Degrees of Freedom': [len(lr_4.params) - 1, df_residual, df_total],
              'Sum of Squares': [explained_ss, residual_ss, total_ss],
              'Mean Square': [ms_explained, ms_residual, None],
              'F-statistic': [f_statistic, None, None],
              'p-value': [p_value, None, None]}

anova_table = pd.DataFrame(anova_data)

# Print the ANOVA table
print("ANOVA Table:")
print(anova_table)
```

```python
# Calculate the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

y_train_price = lr_4.predict(X_train_lm)
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
df_test[num_vars] = scaler.transform(df_test[num_vars])

df_test
```

```python
y_test = df_test.pop('price')
X_test = df_test

# Adding constant variable to test dataframe
X_test_m4 = sm.add_constant(X_test)

# Creating X_test_m4 dataframe by dropping variables from X_test_m4
X_test_m4 = X_test_m4.drop(["bedrooms", "semi-furnished", "basement"], axis = 1)

# Making predictions using the final model
y_pred_m4 = lr_4.predict(X_test_m4)
```

```python
pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred_m4,'Difference':y_test-y_pred_m4})
pred_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(((y_pred_m4 - y_test) ** 2).mean())

print("Root Mean Squared Error (RMSE):", rmse)

# Create a scatter plot of actual vs. predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_m4, alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='r')
plt.title("Actual vs. Predicted Values")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred_m4)

print(f"Mean Squared Error: {mse:.2f}")

# Calculate R-squared score
r2 = r2_score(y_true=y_test, y_pred=y_pred_m4)

print(f"R-squared Score: {r2:.2f}")
```

```python
Accuracy=r2_score(y_test,y_pred_m4)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```
